#### Develop a Python Program called Pattern Generator & Number Analyzer.

```python
print("Welcome to the Pattern Generator & Number Analyzer") 
```

```python
print("Select an Option : ")    # creating a menubar for user to select options
while True :               # using while loop to perform different tasks for diferent inputs
    print("1.Generate pattern")       
    print("2.Analyze Range of Numbers")
    print("3.Exit")
    choice=int(input("Enter your choice : ")) # taking user input of choice
    if choice == 1 :
        a = int(input("Enter the no of rows for pattern : "))
        for i in range(1,a+1) :
            print("*"*i)
        print(sep=" ")
    elif choice == 2:
        no_1 = int(input("Enter the starting range"))
        no_2 = int(input("Enter the ending range"))
        print(sep=" ")
        sum=0
        for i in range(no_1,no_2+1) :
            if i%2==0 :
                print(f"{i} is even")
            else :
                print(f"{i} is odd")
            sum=sum+i
        print(sep="\n")
        print(f"Sum of all numbers from {no_1} to {no_2} is : {sum}")
        print(sep=" ")
    elif choice == 3 :
        print("Exiting the Program. Good Bye!")
        print(sep=" ")
        break
    else :
        print("Enter proper input...")
            
            
```
